```python

df.name.value_counts()
```

# 2.2 Brand having highest no. of outlet

```python
# Get top 10 restaurants
top_10 = df['name'].value_counts().head(10)

# Plot in red
plt.figure(figsize=(8,5))
sns.barplot(x=top_10.values, y=top_10.index, color="red")
plt.xlabel('No. of Outlets')
plt.ylabel('Restaurant')
plt.title('Top 10 Restaurants by Number of Outlets', fontsize=14)
plt.show()
```

# 2.3 Average rating of top 15 outlets

```python
# Average rating of top 15 outlets
avg_rating = df.groupby('name')['aggregate_rating'].mean().sort_values(ascending=False)
# Align with outlet frequency order and take top 15
rating = avg_rating[df['name'].value_counts().index].head(15).round(1)
# Plot in red
plt.figure(figsize=(8,5))
sns.barplot(x=rating.values, y=rating.index, color="red")
plt.xlabel("Average Rating")
plt.ylabel("Outlet")
plt.title("Average Rating of Top 15 Outlets", fontsize=14)
plt.show()
# Reset index (optional, for DataFrame view)
rating.reset_index()
```

# 2.4 Restaurants with high number of votes.

```python
# Top 15 outlets by total votes
vote = df.groupby('name')['votes'].sum().sort_values(ascending=False).head(15)

# Plot in red
plt.figure(figsize=(8,5))
sns.barplot(x=vote.values, y=vote.index, color="red")
plt.xlabel("Total Votes")
plt.ylabel("Outlet")
plt.title("Top 15 Outlets by Total Votes", fontsize=14)
plt.show()

# Display table (optional)
vote
```

# 2.8 Distribution of restaurant's rating according to city

```python
def ci_rating_dist(city_name):
    y = df[df['city'] == str(city_name)]

    plt.figure(figsize=(8,5))
    sns.histplot(y['aggregate_rating'], bins=20, kde=True, color="red")

    plt.title(f"Distribution of Restaurant Ratings in {city_name}", fontsize=14)
    plt.xlabel("Rating")
    plt.ylabel("Count")
    plt.show()
    # choose any city ''ci_rating_dist('here')''
ci_rating_dist('Jaipur')
```

# 2.9 most popular cusins according to restaurants

```python
def cu_res(restaurant_name):
    cuisine = df[df['name'] == restaurant_name]
    plot = cuisine['cuisines'].value_counts().head(10)

    plt.figure(figsize=(8,5))
    sns.barplot(x=plot.values, y=plot.index, color="red")
    plt.title(f"Most Popular Cuisines in {restaurant_name}", fontsize=14)
    plt.xlabel("Number of Restaurants")
    plt.ylabel("Cuisine")
    plt.show()

# Example
cu_res("Barbeque Nation")
```

# 2.20 Word cloud based on customer reviews

```python
from wordcloud import WordCloud
# Combine all rating_text into one string
text = ' '.join(df['rating_text'].astype(str))

# Generate WordCloud in red theme
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='Reds'  # 🔴 red color theme
).generate(text)

# Plot the WordCloud
plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Customer Reviews", fontsize=14)
plt.show()
```
